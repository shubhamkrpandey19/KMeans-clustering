```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
training=pd.read_csv("/content/drive/MyDrive/Nitrile Dataset/SubFP133_Nitrile.csv")
# training.head()
df = training
df.head(5)
```

```python
X=training.iloc[:,5:16097].values
```

```python
from sklearn.cluster import KMeans
```

To get the optimal number of clusters

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init="k-means++", max_iter=300, n_init=10, random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.show()
  
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
train_cluster = kmeans.predict(X)
```

```python
training.insert(1,"Cluster",train_cluster)
```
